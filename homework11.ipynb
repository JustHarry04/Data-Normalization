```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("cars.csv")
```

```python
df_price = df["Price"]
```

```python
df_price = df_price.iloc[0:100]
```

```python
sns.set()
sns.histplot(df_price, kde=True)
plt.show()
```

```python
df_price.describe()
```

## standardisation


```python
standart_price = (df_price-df_price.mean())/df_price.std()
standart_price
```

```python
standart_price.describe()
```

```python
sns.set()
sns.histplot(standart_price, kde=True)
plt.show()
```

# MinMaxScaling

```python
min_max_scale = (df_price-df_price.min())/(df_price.max()- df_price.min())
min_max_scale.describe()
```

```python
sns.set()
sns.histplot(min_max_scale, kde=True)
plt.show()
```

# Mean normalization

```python
mean_normal = (df_price-df_price.mean())/(df_price.max()-df_price.min())
mean_normal.describe()
```

```python
sns.set()
sns.histplot(mean_normal, kde=True)
plt.show()
```

# Maximum Absolute scaling

```python
max_abs_scale = df_price/abs(df_price.max())
max_abs_scale.describe()
```

```python
sns.set()
sns.histplot(max_abs_scale, kde=True)
plt.show()
```

# Scaling to Median and IQR

```python
IQR_scale=(df_price-df_price.median())/(df_price.quantile(0.75)-df_price.quantile(0.25))
IQR_scale.median()
```

```python
sns.set()
sns.histplot(IQR_scale, kde=True)
plt.show()
```

# All in one

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10))  

sns.histplot(df_price, kde=True, color='orange', ax=axs[0, 0])
axs[0, 0].set_title('Car Price Original')
axs[0, 0].set_xlabel('Car Price')
axs[0, 0].set_ylabel('Frequency')

sns.histplot(standart_price, kde=True, color='blue', ax=axs[0, 1])
axs[0, 1].set_title('Standart Car Price')
axs[0, 1].set_xlabel('Car Price')
axs[0, 1].set_ylabel('Frequency')

sns.histplot(min_max_scale, kde=True, color='blue', ax=axs[0, 2])
axs[0, 2].set_title('MinMaxScaled Car Price')
axs[0, 2].set_xlabel('Car Price')
axs[0, 2].set_ylabel('Frequency')

sns.histplot(mean_normal, kde=True, color='blue', ax=axs[1, 0])
axs[1, 0].set_title('Mean Normalized Car Price')
axs[1, 0].set_xlabel('Car Price')
axs[1, 0].set_ylabel('Frequency')

sns.histplot(max_abs_scale, kde=True, color='blue', ax=axs[1, 1])
axs[1, 1].set_title('MaxABS Scaled Car Price')
axs[1, 1].set_xlabel('Car Price')
axs[1, 1].set_ylabel('Frequency')

sns.histplot(IQR_scale, kde=True, color='blue', ax=axs[1, 2])
axs[1, 2].set_title('IQR Scaled Car Price')
axs[1, 2].set_xlabel('Car Price')
axs[1, 2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
